```
import matplotlib.pyplot as plt
```

```
plt.figure(figsize=(10,5))
```

```
plt.figure(figsize=(10,5))

counts = df['AgeofDomain'].value_counts()

plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title("Distribution of Age of Domain")
plt.show()
```

```
plt.figure(figsize=(10,5))

counts = df['AgeofDomain'].value_counts()

plt.bar(counts.index, counts.values, color='skyblue')
plt.xlabel("Age of Domain")
plt.ylabel("Count")
plt.title("Distribution of Age of Domain")
plt.xticks(rotation=45)
plt.show()
```

```
plt.figure(figsize=(10,5))

counts = df['AgeofDomain'].value_counts().sort_index()

plt.plot(counts.index, counts.values, marker='o', linestyle='-', color='green')
plt.xlabel("Age of Domain")
plt.ylabel("Count")
plt.title("Distribution of Age of Domain")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```
import matplotlib.pyplot as plt

counts = df['AgeofDomain'].value_counts()
counts_sorted = counts.sort_index()


plt.figure(figsize=(20,6))

plt.subplot(1, 3, 1)
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Pie Chart: Age of Domain Distribution")


plt.subplot(1, 3, 2)
plt.bar(counts.index, counts.values, color='skyblue')
plt.title("Bar Chart: Age of Domain Distribution")
plt.xlabel("Age of Domain")
plt.ylabel("Count")
plt.xticks(rotation=45)


plt.subplot(1, 3, 3)
plt.plot(counts_sorted.index, counts_sorted.values, marker='o', color='green')
plt.title("Line Chart: Age of Domain Distribution")
plt.xlabel("Age of Domain")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.savefig("dashboard.png", dpi=300, bbox_inches='tight')
plt.show()

```
